```python
import pandas as pd
```

```python
key_df = pd.read_csv("key-countries-pivoted.csv")
```

```python
# Add a total column summing all countries
countries = ['China', 'US', 'United_Kingdom', 'Italy', 'France', 'Germany', 'Spain', 'Iran']
key_df['Total'] = key_df[countries].sum(axis=1)

for column in countries + ['Total']:
    key_df[f'{column}_daily'] = key_df[column].diff()

key_df.loc[0, [f'{column}_daily' for column in countries + ['Total']]] = key_df.loc[0, countries + ['Total']].values

for column in countries + ['Total']:
    key_df[f'{column}_daily_7d'] = key_df[f'{column}_daily'].rolling(window=7).mean().fillna(0).round().astype(int)

key_df.drop(columns=countries, inplace=True)

# Save the updated DataFrame to a new CSV file
key_df.to_csv("key-countries-pivoted.csv", index=False)

print(key_df.head())
```

```python
import os
key_df['Date'] = pd.to_datetime(key_df['Date'])

for country in countries:
    folder_name = country.lower().replace('_', '-')
    os.makedirs(folder_name, exist_ok=True)
    country_df = key_df[['Date', f'{country}_daily', f'{country}_daily_7d']]
    
    country_df = country_df.rename(columns={
        f'{country}_daily': 'daily_cases',
        f'{country}_daily_7d': 'daily_cases_7d_avg'
    })

    filename = f"{folder_name}/daily_cases.csv"
    country_df.to_csv(filename, index=False)
    
    print(f"Saved {filename}")
```
